```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
plt.figure(figsize=(12,10))
sns.set(style='darkgrid')
ax = sns.countplot(x=df.rating, data=df, palette='Set2', order=df.rating.value_counts().index[0:15])
```

```python
imdb_ratings = pd.read_csv("IMDb ratings.csv")
imdb_ratings.head()
```

```python
imdb_movies = pd.read_csv('IMDb movies.csv')
imdb_movies.head()
```

```python
imdb_movies.country.value_counts()
```

```python
plt.figure(figsize=(12,8))
sns.set(style='darkgrid')
ax = sns.countplot(x=imdb_movies.country, data=imdb_movies, order=imdb_movies.country.value_counts().index[0:5])
```

```python
from math import floor
ratings = []
for i in imdb_ratings.weighted_average_vote:
    ratings.append(floor(i))
```

```python
sns.set(style='darkgrid')
ax = sns.countplot(y=ratings, data=imdb_ratings)
ax.set_ylabel('ratings')
```
