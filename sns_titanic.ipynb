```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df2 = df.drop('PassengerId',axis=1)
df2.head()
```

```python
df2.Pclass.value_counts()
```

```python
sns.relplot(x=df2.Sex,y=df2.Age, hue=df2.Pclass, data=df2)
```

```python
sns.relplot(x=df2.Fare, y=df2.Age, hue=df2.Sex, style=df2.Pclass, data=df2, palette='ch:r=0.3, l=0.9')
```
